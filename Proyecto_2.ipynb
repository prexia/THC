```python
import pandas as pd #Importamos la biblioteca pandas para poder trabajar con el base de datos
import matplotlib.pyplot as plt #Importamos la bibliota de matplotlib para poder ingresar graficas
```

```python
link='https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv'
link='https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv'
#Agregamos links de la base de datos donde vamos a sacar la información
code='?raw=true'
url=link+code
df = pd.read_csv(url,sep=',',index_col=1)
```

```python
df #Imprimimos la informacion
```

```python
df.tail()
```

```python
cuantos=df.groupby('Country').size().reset_index(name='Contabilizados')
cuantos
```

```python
USA=df[df['Country']=='United States of America'] #Definimos a Estados Unidos como USA
```

```python
USA=USA.drop('Country',axis=1)
```

```python
plt.pie(cuantos['Contabilizados'],labels=cuantos['Country'],autopct='%0.1f%%') #Hacemos una grafica de pay de los paises de la lista de cuantos
plt.axis('equal')
plt.title("Distribucion del ebola") #Titulo de la grafica
plt.savefig("Pastel_Paises.png",dpi=300,format='png') #dpi=Puntos por pixeles Guardamos la imgaen
plt.show()
```

```python
USA.isnull().sum()
```

```python
mean_values=USA['No. of confirmed deaths'].mean() #Los valores de las muentres confirmandas en USA
mean_values
```

```python
Datos=USA.index[0:-1:10] #Definimos los datos de USA del -1 al 10
NCC=USA['No. of confirmed cases'][0:-1:10] #Definimos NCC como los casos confirmados
NCD=USA['No. of confirmed deaths'][0:-1:10] #Definimos NCC como los casos muertos
```

```python
plt.plot(Datos,NCC,'--k',lw=4,label=' Personas Infectadas')#Insertamos la linea de casos confirmados
plt.plot(Datos,NCD,':b',lw=4,label='Decesos')#Insertamos la linea de casos muertos
plt.title('Avance del ébola en Estados Unidos',color='red') #Titulo de la Grafica
plt.ylabel('Casos confirmados') #Nombre del eje y
plt.xlabel('Periodo de Tiempo') #Nombre del eje x
plt.xticks(rotation=90,fontsize=8) #rotacion y tamaño de los valores fuera de la grafica
plt.grid() #Grilla
plt.legend(loc='upper left',bbox_to_anchor=(1.02,0.8)) #Cuadro de informacion en la grafica
plt.savefig("Grafica de barras sin barras ebola USA.png",dpi=300,format='png') #Guardar la grafica
plt.show()
```

```python
fig=plt.figure(figsize=(12,8))#Insertar figura y tamaño
plt.plot(Datos,NCC,'b',label='Infectados') #Insertamos la linea de datos infectados
plt.bar(Datos,NCC,color='yellow')#Insertamos la barra con los datos infectados
plt.plot(Datos,NCD,'r',label='Decesos')#Insertamos la linea de datos de decesos
plt.bar(Datos,NCD,color='orange')#Insertamos la barra con los datos de decesos
plt.grid() #Grilla
plt.title('Avance del ebola en Estados Unidos ',color='red') #Titulo de la gráfica
plt.ylabel('Casos confirmados') #Titulo del eje y
plt.xlabel('Periodod de Tiempo') #Titulo del eje x
plt.xticks(rotation=90,fontsize=9) #rotacion y tamaño de los valores fuera de la grafica
plt.legend(loc='upper left',bbox_to_anchor=(1.02,0.8))#Cuadro de informacion en la grafica
plt.savefig("Grafica de barras ebola USA.png",dpi=300,format='png')#Guardar la grafica
plt.show()
```

```python
PIvsD=100*USA['No. of confirmed deaths']/USA['No. of confirmed cases'] # Definimos el número de datos de muertos menos los casos confirmados
USA['PIvsD']=PIvsD #En USA lo definimos como PIvsD
```

```python
PIvsD
```

```python
USA.head()
```

```python
USA['PR']=(100-PIvsD) #PR en USA igual a 100-menos los datos de PIvsD
```

```python
PIvsD_mean=USA['PIvsD'].mean()#Los datos de PIvsD
PR_mean=USA['PR'].mean()#Los datos de PR
porcentajes=[PIvsD_mean,PR_mean] #Porcentajes igual a los datos de PIvsD y PR
etiquetas=['PIvs','PR'] #Definimos a las etiquetas con sus respectivos nombres
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct='%0.i%%',colors=['yellow','green'])#Insertamos grafica de pay
plt.axis("equal")
plt.title("Porcentaje recuperado contra decesos")#Titulo de la grafica
plt.savefig("Recuperados de ebola USA.png",dpi=300,format='png')#Guardamos la figura
plt.show()
```

```python
G=USA['Girls'][0:-1:15] #Definimos a G como las niñas en Estados Unidos
AF=USA['Adult Female'][0:-1:15]#Definimos a AF como las adultas en Estados Unidos
OF=USA['Old Female'][0:-1:15]#Definimos a OF como las adultas mayores en Estados Unidos
Datos=USA.index[0:-1:15] #Definimos el rango de los datos
fig=plt.figure(figsize=(15,10)) #Insertamos la figura y su tamaño
plt.bar(Datos,AF,color='blue',alpha=0.8,label='Niñas')#Insertamos la barra de las niñas con los datos ya asignados
plt.bar(Datos,G,color='black',alpha=0.8,label='Adultas')#Insertamos la barra de las adultas con los datos ya asignados
plt.bar(Datos,OF,color='red',alpha=0.8,label='Adultas mayores')#Insertamos la barra de las adultas mayores con los datos ya asignados
plt.grid() #Grilla
plt.title("Distribución del ebola en mujeres en Estados Unidos") #Titulo de la grafica
plt.xticks(rotation=90,fontsize=8)#rotacion y tamaño de los valores fuera de la grafica
plt.legend(loc='upper left',bbox_to_anchor=(1.02,0.5))#Cuadro de informacion en la grafica
plt.savefig("Grafica de mujeres con ebola USA.png",dpi=300,format='png')#Guardar la grafica
plt.show()
```

```python
B=USA['Boys'][0:-1:15]#Definimos a los niños de Estados Unidos con una B
AM=USA['Adult Male'][0:-1:15]#Definimos a los adultos de Estados Unidos con una AM
OM=USA['Old Male'][0:-1:15]#Definimos a los adultos mayores de Estados Unidos con una OM
Datos=USA.index[0:-1:15]#Definimos el rango de los datos
fig=plt.figure(figsize=(15,10)) #Insertamos la figura y su tamaño
plt.bar(Datos,AM,color='blue',alpha=0.8,label='Niños')#Insertamos la barra de los niños con los datos ya asignados
plt.bar(Datos,B,color='yellow',alpha=0.6,label='Adultos')#Insertamos la barra de los adultos con los datos ya asignados
plt.bar(Datos,OM,color='black',alpha=0.4,label='Adultos mayores')#Insertamos la barra de los adultos mayores con los datos ya asignados
plt.grid()#Grilla
plt.title("Distribución del ebola en Hombres en los Estados Unidos")#Titulo de la grafica
plt.xticks(rotation=90,fontsize=8)#rotacion y tamaño de los valores fuera de la grafica
plt.legend(loc='upper left',bbox_to_anchor=(1.02,0.5))#Cuadro de informacion en la grafica
plt.savefig("Grafica de hombres con ebola USA.png",dpi=300,format='png')#Guardar la grafica
plt.show()
```
