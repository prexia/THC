```python
SL=df[df['Country']=='Sierra Leone']
```

```python
SL.isnull().sum()
```

```python
mean_values=SL['No. of confirmed deaths'].mean()
mean_values
```

```python
SL.fillna(mean_values,inplace=True)
```

```python
Dates=SL.index[0:-1:10]
NCC=SL['No. of confirmed cases'][0:-1:10]
NCD=SL['No. of confirmed deaths'][0:-1:10]
```

```python
plt.plot(Dates,NCC,'o--k',label='Infectados')
plt.plot(Dates,NCD,'D--b',label='Decesos')
plt.grid()
plt.title('Avance del ebola en Sierrra Leona ')
plt.ylabel('Casos confirmados')
plt.xlabel('Periodod de Tiempo')
plt.xticks(rotation=90,fontsize=9)
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(12,8))
plt.plot(Dates,NCC,'*--b',label='Infectados')
plt.bar(Dates,NCC)
plt.plot(Dates,NCD,'D-r',label='Decesos')
plt.bar(Dates,NCD)
plt.grid()
plt.title('Avance del ebola en Sierrra Leona ')
plt.ylabel('Casos confirmados')
plt.xlabel('Periodod de Tiempo')
plt.xticks(rotation=90,fontsize=9)
plt.legend()
plt.savefig("Grafica de barras _ebola.png",dpi=300,format='png')
plt.show()
```

```python
PIvsD=100*SL['No. of confirmed deaths']/SL['No. of confirmed cases']
SL['PIvsD']=PIvsD
```

```python
SL.head()
```

```python
recuperados=(100-PIvsD/100)*SL['No. of confirmed cases']
SL['recuperados']=recuperados
```

```python
SL.head()
```

```python
SL['PR']=(100-PIvsD)
```

```python
SL.head()
```

```python
PIvsD_mean=SL['PIvsD'].mean()
PR_mean=SL['PR'].mean()
porcentajes=[PIvsD_mean,PR_mean]
etiquetas=['PIvs','PR']
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct='%0.i%%',colors=['yellow','magenta'])
plt.axis("equal")
plt.title("Porcentaje recuperado contra decesos")
plt.show()
```

```python
Boys=SL['Boys'].to_numpy()
SL_Boys=pd.DataFrame({'Boys':Boys})
SL_Boys.index=pd.to_datetime(SL.index,format='%d/%m/%Y')
SL_Boys=SL_Boys.sort_index()
SL_Boys
```

```python
steps=57
data_train=SL_Boys[:-steps]
data_test=SL_Boys[:-steps]
print(f"Las fechas son {len(data_train)}")
```

```python
fig,ax=plt.subplots(figsize=(8,3))
data_train['Boys'].plot(ax=ax,color='blue',label='train')
data_test['Boys'].plot(ax=ax,color='red',label='test')
ax.legend()
plt.show()
```

```python
from skforecast.ForecasterAutoreg import ForecasterAutoreg
from sklearn.ensemble import RandomForestRegressor
```

```python
from numpy import random
forecaster=ForecasterAutoreg(
  regressor=RandomForestRegressor(random_state=123),
  lags=6
)
```

```python
forecaster.fit(y=data_train['Boys'])
```

```python
steps=57
predicciones=forecaster.predict(steps=steps)
```

```python
predicciones.index=data_test.index
```

```python
fig,ax=plt.subplots(figsize=(8,3))
data_train['Boys'].plot(ax=ax,color='blue',label='train')
data_test['Boys'].plot(ax=ax,color='red',label='test')
predicciones.plot(ax=ax,color='black',label='predicciones')
ax.legend()
plt.show()
```
