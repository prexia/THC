Ejercicio 1: Con base en la definición de número primo, escriba una función que reciba un número entero e imprima 'True' si es primo o de lo contrario imprima 'False'. Tenga en cuenta que si el número es menor o igual a 1, se retorna como resultado "Error en el Número" y debe volver a preguntar hasta que la entrada sea valida.

```python
#Por definicion de primo tenemos que un numero es primo si solo es divisible entre 1 y si mismo y el residuo es 0
def funcion_primos(Numero): #Definimos una funcion llamada funcion_primos la cual va a evaluar el numero
  if Numero<=1: #Si el numero es menor que el 1
    return("Error de numero") #Regrese error de numero
  elif Numero>1: #Ahora si el numero es mayor que 1
      for x in range(2,int(Numero**0.5)+1): #Para cada x en el rango de los numeros enteros si el numero >1 al momento de multiplicarlo por **5 y sumarle 1
        if (Numero % x)==0: #Si ese numero "modificado" es divisible entre si mismo y da 0
            return "Falso" #Regrese falso
      else: #Sino
          return "True"#Regrese verdadero

num=float(input("Ingrese un numero ")) #num= que el usuario agregue el numero que gusta
primo=(funcion_primos)(num) #primo= el numero q ingresamos ya evaluado con la funcion
print(primo) #Imprimimos "primo"
```

Ejercicio 2:Sin utilizar librerías adicionales cree un arreglo de numéros enteros (entre 0 y 150) y calcule a través de funciones estándar o lambda lo siguiente:

El promedio
La varianza
La desviación estándar
Visualice los datos en un histograma.
La suma de los números pares e impares.
Los múltiplos de 2, 3 y 5.

```python
#1 Promedio
numeros=(1,150)#Lista de numeros
promedio=sum(numeros)/len(numeros)#Promedio igual a la suma de los valores en numeros/ la cantidad de valores que hay
print(f"El promedio de la suma de los numeros entre 1 y 150 es igual a :{promedio} ") #Imprimir promedio
#Con funciones
def fun_promedio(numeros): #Definimos una funcion
    suma = sum(range(1,151)) #Definimos la suma como la suma de todos los valores en el rango de 1 a 151 ya que al 150 va a tomar hasta el 149
    return suma / len(range(1,151)) # Que devuelva la suma / la cantidad de valores en el ranfo de 1 a 151

promedio = fun_promedio(numeros) #Definimos al promedio como la funcion ya establecida
print(f"El promedio de la suma de los numeros entre 1 y 150 es igual a :{promedio} ")#Imprimimos el promedio
```

```python
# 2 La varianza
Valores=(range(1, 151))# Lista de números
promedio = sum(Valores) / len(Valores)# Calcular el promedio
#Por definicion de varianza tenemos que calcular la diferencia el cuadrado de cada numero en particular
#Para cada x en Valores
diferencias_cuadrado = [(x - promedio) ** 2 for x in Valores ]#Diferencias del cuadrado es igual a la x menos el promedio ya calculado al cuadrado
varianza = sum(diferencias_cuadrado) / len(diferencias_cuadrado) #Variaza= La suma de las diferencias al cuadrado entre la cantidad de elementos de las diferencias del cuadrado

print(f"La varianza de la lista de numeros que va del 1 al 150 es: {varianza}")
```

```python
# 3 La Desviacion Estandar
Valores=(range(1, 150)) #Lista de Numeros hasta el 151 pq si no el 150 no lo toma
arreglo_graf = np.array(Valores)# Convertir la lista a un arreglo de la biblioteca numpy
des_estandar = np.std(arreglo_graf)# Calcular la desviación estándar usando numpy
print(f"La Desviación estándar:{des_estandar}")#Imprimimos la Desviacion estandar
```

```python
#3.1 Visualice los datos en un histograma
Valores=(range(1,150))#Definimos la lista de numeros del 1 al 150n
plt.hist(Valores, bins=10, color='Red', edgecolor='black',lw=2,linestyle='--') #Ocupamos el plt.hist para sacar un hsitograma y lo modificamos el color,tipo de linea,etc.
plt.title('Histograma de los datos obtenidos')#Nombre del histograma
plt.xlabel('Valores')#Nombre del eje x
plt.ylabel('Frecuencia')#Nombre del eje y
plt.grid()#Grilla
plt.show()
```

```python
# 5 Suma de numeros pares e imprares
Valores=(range(1,150)) # Lista de numeros
sumadepares = 0 #Definimos la suma de los pares = 0 para que no sume valores extra
sumadeimpares = 0 #Definimos la suma de los impares = 0 para que no sume valores extra
for x in Valores: #Para cada x in valores
    if x % 2 == 0: # Si x es divisible entre 2 y da de residuo 0
        sumadepares += x #Suma de pares +x
    else:
        sumadeimpares += x #Si al dividirlo entre 2 no da de residuo 0 sumarlo con los impares

# Imprimir los resultados
print(" La suma de valores pares en el rango de 1 a 150 es ", sumadepares) #Imprimir la suma de los numeros pares
print("La suma de valores pares en el rango de 1 a 150 es ", sumadeimpares) #Imprimir la suma de los numeros impares
```

```python
# 6 Los multiplos de 2,3,5
Valores=range(1,151) #Lista de numeros

print("Los Multiplos de 2 son: ") #Imprimir los multiplos de 2 si
for x in Valores: # Para cada x en Valores
  if x % 2 == 0: #Si al dividirlo entre 2 y da residuo 0
    print(x) #Imprimir x

print("Los Multiplos de 3 son: ") #Imprimir los multiplos de 3 si
for y in Valores:# Para cada y en Valores
  if y % 3 == 0: #Si al dividirlo entre 3 y da residuo 0
    print(y) #Imprimir y

print("Los multiplos de 5 son: ")#Imprimir los multiplos de 5 si
for z in Valores:#Para cada z en Valores
  if z % 5 == 0:#Si al dividirlo entre 5 y da residuo 0
    print(z) #Imprimir z

print("Los multiplos de 2,3,5 son: ")#Imprimir los multiplos de 2,3,5 si
for j in Valores:#Para cada j en Valores
  if j % 30 == 0:#Si al dividirlo entre 2,3,5 y da residuo 0
    print(j)#Imprimir j
```

```python
fig=plt.figure(figsize=(10,8))#Tamaño de la figura
x=np.array([])#Arreglo de x
y=np.array([])#Arreglo de y
c=np.array([])#Arreglo de c
def curvas(x, C): #Definimos una funcion la cual nos va a hacer que el valor c camcie en base a los valores de x
    return (x - 1) + C * np.exp(-x) #Y nos va a devolveruna curva
posibles_valores = np.linspace(-2, 5, 10) #Estos son los posibles valores de x en los cual se va a definir C
C_mayor_0 = [1, 2, 3] #Definimos algunos valores no tan grandes para cuando C sea mayor que 0
C_igual_0 = [0]#Definimos para cuando C sea igual que 0
C_menor_0 = [-1, -2, -3]#Definimos algunos valores no tan grandes para cuando C sea menor que 0
for c in C_mayor_0:#Para algun c que pretenezca al arreglo que hicimos donde C>0
    valores_y = curvas(posibles_valores, c)#Se  le aplica la funcion a ese c
    plt.plot(posibles_valores, valores_y, label='C>0',lw=3, linestyle='--') #De tal forma que nos devuelve cada una de las curvas graficadas en los valores,xy,

for C6 in C_igual_0:#Para algun C6 que pretenezca al arreglo que hicimos donde C=0
    valores_y = curvas(posibles_valores, C6)#Se  le aplica la funcion a ese C6
    plt.plot(posibles_valores, valores_y, label='C=0',lw=3, linestyle='-',)#De tal forma que nos devuelve cada una de las curvas graficadas en los valores,xy,

for j in C_menor_0:#Para algun j que pretenezca al arreglo que hicimos donde C<0
    valores_y = curvas(posibles_valores, j)#Se  le aplica la funcion a ese j
    plt.plot(posibles_valores, valores_y, label='C<0',lw=3, linestyle=':',)#De tal forma que nos devuelve cada una de las curvas graficadas en los valores,xy,

plt.plot(x,y,lw=2.5,label="Cuando C<0 la funcion se ve asi")#Graficacion de de la ecuacion cuando c es menor que 0
x=([-5,-4,-2,6,15,20])#Valores de x
y=([25,11,1,5,14,19])#Valores de y
plt.plot(x,y,lw=3,label="Cuando C>0 la funcion se ve asi")#Graficacion de de la ecuacion cuando c es mayor que 0
x=np.array([0,-3,-19])#Valores de x
y=np.array([-1,-4,-20])#Valores de y
plt.plot(x,y,lw=2.5,label="Cuando C=0 la funcion se ve asi")#Graficacion de de la ecuacion cuando c es igual que 0
plt.title(r'$Grafica\; de\; los\; distintos\; valores; de\; C\; en\; la\; ecuacion:y=(x-1)+Ce^{-x}\;y\;su\;familia\;de\;curvas$')#Titulo de la grafica
plt.grid()#Grilla en la grafica
plt.subplot(1,1,1)#Cantidad de columnas y renglones
plt.ylabel('Eje y')#Nombre del eje y
plt.xlabel('Eje x')#Nombre del eje x
plt.xticks(fontsize=8)#Tamaño de las coordenadas del eje x
plt.yticks(fontsize=8)#Tamaño de las coordenadas del eje y
plt.legend(loc='upper left',bbox_to_anchor=(1.02,0.8))#Coordenadas y tamaño del cuadro de etiqueta de la gráfica
plt.show()
```
